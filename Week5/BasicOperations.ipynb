```python
mImage = cv2.imread('unicorn.jpg')

hsvImg = cv2.cvtColor(mImage,cv2.COLOR_BGR2HSV)

# decreasing the V channel by a factor from the original
hsvImg[...,2] = hsvImg[...,2]*0.6

plt.imshow(cv2.cvtColor(hsvImg,cv2.COLOR_HSV2RGB))
plt.title('darkened image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
mImage2 = cv2.imread('dark.jpg')

hsvImg2 = cv2.cvtColor(mImage2,cv2.COLOR_BGR2HSV)

# decreasing the V channel by a factor from the original
hsvImg2[...,2] = hsvImg2[...,2]*1.2

plt.imshow(cv2.cvtColor(hsvImg2,cv2.COLOR_HSV2RGB))
plt.title('brighter image'), plt.xticks([]), plt.yticks([])
plt.show()

plt.imshow(cv2.cvtColor(mImage2,cv2.COLOR_BGR2RGB))
plt.title('original image'), plt.xticks([]), plt.yticks([])
plt.show()

hsvImg3 = cv2.cvtColor(mImage2,cv2.COLOR_BGR2HSV)
hsvImg3[...,2] = hsvImg3[...,2]*0.6
plt.imshow(cv2.cvtColor(hsvImg3,cv2.COLOR_HSV2RGB))
plt.title('darker image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blur_image = cv2.GaussianBlur(img, (55,55),0)
#cv2.imshow('blurred image',blurImg) 
plt.imshow(cv2.cvtColor(blur_image,cv2.COLOR_BGR2RGB))
#plt.imshow(blur_image)
plt.title('blur image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Canny edge detection
canny_image = cv2.Canny(img, 55,55)
#cv2.imshow('blurred image',blurImg) 
#plt.imshow(cv2.cvtColor(canny_image,cv2.COLOR_BGR2RGB))
plt.imshow(canny_image)
plt.title('canny edge'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Blur followed by edge
blur_image2 = cv2.GaussianBlur(img, (15,15),0)
#cv2.imshow('blurred image',blurImg) 

canny_image2 = cv2.Canny(blur_image2, 55,55)
plt.imshow(canny_image2)
plt.title('canny edge with blur'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
canny_image = cv2.Canny(mImage2, 55,55)
plt.imshow(canny_image)
plt.title('canny edge'), plt.xticks([]), plt.yticks([])
plt.show()
blur_image3 = cv2.GaussianBlur(mImage2, (17,17),0)

canny_image3 = cv2.Canny(blur_image3, 55,55)
plt.imshow(canny_image3)
plt.title('canny edge with blur'), plt.xticks([]), plt.yticks([])
plt.show()
```
